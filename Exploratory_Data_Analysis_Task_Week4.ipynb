## Q.1:Import libraries (Numpy, pandas, matplotlib, plotly and seaborn) and then read csv file.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

df = pd.read_csv("menu.csv")

df.head()
```

## Q.2: Check statistical facts by checking all columns. Then calculate the maximum value of the following attributes.

```python

statistical_facts = df.describe()
print("Statistical Facts:")
print(statistical_facts)

attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars',
              'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
              'Calcium (% Daily Value)', 'Iron (% Daily Value)']

max_values = df[attributes].max()
print("\nMaximum Values:")
print(max_values)
```

## Q.3: Check to see if infact there is any correlation between Calories and other independent variables by plotting a correlation matrix next.

```python

correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

## Q.4: Draw boxplot for Calories vs Category to spot outliers and max calories category.

```python

plt.figure(figsize=(12, 8))
sns.boxplot(x='Calories', y='Category', data=df)
plt.title('Category vs Calories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Q.5: Figure out which exact item contains a high quantity for attributes and display summary.

```python

attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']

summary = {}
for attribute in attributes:
    max_value = df[attribute].max()
    item_with_max_value = df.loc[df[attribute].idxmax(), 'Item']
    summary[attribute] = {'Item': item_with_max_value, 'Max Value': max_value}

for attribute, values in summary.items():
    print(f"Attribute: {attribute}")
    print(f"Item with Max Value: {values['Item']}")
    print(f"Max Value: {values['Max Value']}")
    print()
```

## Q.6: Draw Stripplot for each category against the following attributes.

```python

attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars',
              'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
              'Calcium (% Daily Value)', 'Iron (% Daily Value)']

for attribute in attributes:
    plt.figure(figsize=(12, 8))
    sns.stripplot(x='Category', y=attribute, data=df, jitter=True)
    plt.title(f'{attribute} by Category')
    plt.xticks(rotation=45, ha='right')
    plt.xlabel('Category')
    plt.ylabel(attribute)
    plt.show()
```

## Q.7: Draw a horizontal bar graph for items in each category against the calories.

```python

grouped_data = df.groupby('Category')

for category, data in grouped_data:

    plt.barh(data['Item'], data['Calories'])

    plt.xlabel('Calories')
    plt.ylabel('Items')
    plt.title(f'Calories per Item - Category: {category}')

    plt.show()
```
